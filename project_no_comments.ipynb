## Описание проекта

Необходимо подготовить прототип модели машинного обучения с учителем в качестве решения для эффективной работы промышленных предприятий.

Модель должна будет предсказать коэффициент восстановления золота из золотосодержащей руды на некотором производстве. Имеются данные с параметрами добычи и очистки. 

Данные находятся в трёх файлах:

- gold_recovery_train_new.csv — обучающая выборка;
- gold_recovery_test_new.csv — тестовая выборка;
- gold_recovery_full_new.csv — исходные данные.

Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.

Описание данных:

- Rougher feed — исходное сырье;
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant;
- Rougher process (англ. «грубый процесс») — флотация;
- Rougher tails — отвальные хвосты;
- Float banks — флотационная установка;
- Cleaner process — очистка;
- Rougher Au — черновой концентрат золота;
- Final Au — финальный концентрат золота;

Параметры этапов:

- air amount — объём воздуха;
- fluid levels — уровень жидкости;
- feed size — размер гранул сырья;
- feed rate — скорость подачи.

Названия этапов:

- rougher — флотация;
- primary_cleaner — первичная очистка;
- secondary_cleaner — вторичная очистка;
- final — финальные характеристики.

Возможные типы параметров:

- input — параметры сырья;
- output — параметры продукта;
- state — параметры, характеризующие текущее состояние этапа;
- calculation — расчётные характеристики.

Для решения задачи необходимо использовать метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

Нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата rougher.output.recovery;
- эффективность обогащения финального концентрата final.output.recovery.

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

### Подготовка данных

Преобразуем столбец date во всех данных к требуемуму типу и сделаем его индексом.

```python
# Приводим время к формату DateTime
df_train['date'] = pd.to_datetime(df_train['date'])
df_test['date'] = pd.to_datetime(df_test['date'])
df_full['date'] = pd.to_datetime(df_full['date'])
```

```python
# Устанавливает даты в качестве индекса
df_train = df_train.set_index('date')
df_test = df_test.set_index('date')
df_full = df_full.set_index('date')
```

Проверим наличие полных дубликавто в данных:

```python
# Проверка наличия полных дубликатов в данных
print('Наличие полных дубликатов в исходных данных:', df_full.duplicated().sum())
print('Наличие полных дубликатов в обучающих данных:', df_train.duplicated().sum())
print('Наличие полных дубликатов в тестовых данных:', df_test.duplicated().sum())
```

Убедимся, что recovery rate в обучающей выборке рассчитан верно:

```python
# Расчёт recovery rate для обучающей выборки
au_before = df_train['rougher.input.feed_au']
au_after = df_train['rougher.output.concentrate_au']
au_tail = df_train['rougher.output.tail_au']

au_rec = ((au_after * (au_before - au_tail)) / (au_before * (au_after - au_tail))) * 100
```

```python
# Расчёт MAE для обучающей выборки
rec_mae = mean_absolute_error(au_rec, df_train['rougher.output.recovery'])
round(rec_mae, 3)
```

Recovery Rate был рассчитан верно.

В обучающих данных имеются столбцы, которых нет в тестовых данных. Нет необходимости сохранять в обучающей выборке такие столбцы, так как обученная модель не сможет получить такие данные при тесте. Удалим их:

```python
# Создаёт список из названий столбцов, которые надо удалить в обучающей выборке
col_test = list(df_test.columns)
col_train = list(df_train.columns)
col_to_del = [i for i in col_train if i not in col_test]
# Удаление целевых признаков из списка
col_to_del.remove('final.output.recovery')
col_to_del.remove('rougher.output.recovery')
```

Список признаков, недоступных на тесте:

```python
col_to_del
```

В тестовой выборке недостпуны признаки, которые являются: расчётными, "хвостами", а параметры продукта выхода. Причиной отсутствия данных признаков в тесте может быть то, что в действительности эти данные появляются уже после всего процесса и не могут служить признаками для предсказания (так как неизвестным в момент предсказания).

```python
# Удаление столбцов из train, которых нет в test
df_train = df_train.drop(col_to_del, axis=1)
```

Кроме того в тестовой выборке отсутствуют целевые показатели. Перенесём их из исходных данных по общим индексам времени.

```python
# Перенос целевых показателей из full в test
df_test['final.output.recovery'] = df_full.loc[df_test.index, 'final.output.recovery']
df_test['rougher.output.recovery'] = df_full.loc[df_test.index, 'rougher.output.recovery']
```

Заменим во всех неотрицательных столбцах значения, которые меньше четверти среднеквадратичного отклонения, на np.nan:

```python
# Цикл, который заменяет выбросы на nan по нижней границе (половина среднеквадратичного отклонения)
# для столбцов с неотрицательными значениями
for i in df_test.columns:
    if df_test[i].sum() > 0:
        df_test.loc[df_test[i] < df_test[i].std()*0.25, i] = np.nan
```

Функция расчёта sMAPE, которая будет использоваться в качестве функции оценки потерь:

```python
# Функция расчёта sMAPE
def smape(target, predict):
    smape = (1 / len(target)) * np.sum((np.abs(target - predict)) / ((np.abs(target) + np.abs(predict)) / 2)) * 100
    return round(smape, 3)
```

Функция для расчёта итоговой метрики:

```python
# Итоговая метрика
def res_smape(rough, final):
    smape = 0.25 * rough + 0.75 * final
    return round(smape, 3)
```

```python
# Создание отсортированного списка со значениями F-statistic для признаков
freg_rough = f_regression(features_train, target_train['rougher.output.recovery'])
freg_rough = pd.Series(freg_rough[0], index=features_train.columns).sort_values(ascending=False)

# Цикл для определия лучшего кол-ва признаков
best_result = 100
best_i = 0

for i in range(1, 26):
    freg_rough_best = freg_rough.index[:i] # Срез признаков
    
    # Обучение модели линейной регрессии
    lr_rough = LinearRegression()
    lr_rough_smape = cross_val_score(
                                lr_rough,
                                features_train[freg_rough_best],
                                target_train['rougher.output.recovery'],
                                scoring=smape_score
    )
    result = np.abs(lr_rough_smape).mean()
    
    # Фиксация лучших результатов
    if result < best_result:
        best_result = result
        best_i = i

print('Кол-во признаков:', best_i)        
print('Лучший sMAPE:', round(best_result, 3))

lr_rough_smape = best_result # Фиксация лучшего результата для флотации
```

#### Подбор оптимального количества признаков для финальной очистки

```python
# Создание отсортированного списка со значениями F-statistic для признаков
freg_final = f_regression(features_train, target_train['final.output.recovery'])
freg_final = pd.Series(freg_final[0], index=features_train.columns).sort_values(ascending=False)

# Цикл для определия лучшего кол-ва признаков
best_result = 100
best_i = 0

for i in range(1, 26):
    
    freg_final_best = freg_final.index[:i] # Срез признаков
    
    # Обучение модели линейной регрессии
    lr_final = LinearRegression(fit_intercept=True)
    lr_final_smape = cross_val_score(
                                lr_final,
                                features_train[freg_final_best],
                                target_train['final.output.recovery'],
                                scoring=smape_score
    )
    result = np.abs(lr_final_smape).mean()
    
    # Фиксация лучших результатов
    if result < best_result:
        best_result = result
        best_i = i

print('Кол-во признаков:', best_i)        
print('Лучший sMAPE:', round(best_result, 3))

lr_final_smape = best_result # Фиксация лучшего результата для финальной очистки
```
